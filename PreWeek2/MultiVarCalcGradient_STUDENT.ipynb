## What is a Partial Derivative?
If you are comfortable with taking a derivative of a function of a single variable, then partial derivatives should be pretty straightforward. Suppose I have a function that depends on two variables:

\begin{aligned}
f(x, y) = 2x^{2} + xy
\end{aligned}

A graph of ($f(x,y)$) would look like a sloping valley. We can walk anywhere in the horizontal plane (in the $x$ and $y$ directions), and the value of $f(x,y)$ tells us our altitude at that point on the valley's surface. Now, a typical derivative tells us the slope of our function wherever we are standing. But that is when we can only move along *one* axis (when $f$ depends on *one* variable only). If I ask you: "what is the slope of $f(x,y)$ where you are standing?", you would naturally respond "in which direction do you mean"? Perhaps the valley is flat along the direction you are facing, whereas the valley slopes upward towards your right (and thus downward to your left). Once we are dealing with functions depending on more than one variable, we must specify the *direction* along which we are taking our derivative. This is where partial derivatives come into play.

The <strong>partial derivative of $f$ with respect to $x$</strong>, denoted $\frac{\partial f}{\partial x}$, gives us the slope of $f(x,y)$ <strong>along the $x$-direction</strong>. To accomplish this, you simply take the derivative of $f$ as if $x$ was the  *only* variable that $f$ depended on (i.e. as if $y$ were a constant, like the number 3).   

Let's compute this:

\begin{aligned}
\frac{\partial f(x, y)}{\partial x} = 4x + y
\end{aligned}

just as the derivative of $3x$ is $3$, the partial derivative of the term $xy$, with respect to $x$, is $y$. Notice that $\frac{\partial f}{\partial x}$ is, itself, a function of both $x$ and $y$. This means that the slope of $f$ along the $x$-direction depends on both $x$ and $y$. This should make intuitive sense - if I ask you what the slope of the valley is along the $x$-direction, your answer will depend on *where* in the valley you are standing.

There is nothing special about $x$, we can also ask for the slope of $f(x,y)$ along the $y$-axis (i.e. take the derivative as if $y$ was the only variable):

\begin{aligned}
\frac{\partial f(x, y)}{\partial y} = x
\end{aligned}

The term $2x^{2}$ did not involve $y$ at all, so its partial derivative with respect to $y$ is 0 (just like how the derivative of 2 is 0). It turns out that $\frac{\partial f}{\partial y}$ is only a function of $x$ - I can tell you the slope of $f$ in the $y$ direction by only knowing where I stand along the $x$ axis (it doesn't matter what my $y$-coordinate is). This is merely a feature of this specific function we are working with.

## Taking a Derivative Along *Any* Direction Using the Gradient

Finally, and I can't give this the full treatment it deserves, what if we want to take the derivative of $f(x,y)$ along some *other* direction? Say, we are facing 45 degrees between the $+x$ and $+y$ direction, and we want to know the slope of $f$ along *this* direction. This is where the <strong>gradient</strong> comes into play, which is described in the module. In short, the gradient of $f(x,y)$ is the <strong>vector containing all of the partial derivatives of $f$</strong>:

\begin{aligned}
\nabla f(x,y) = \lbrack \frac{\partial f(x, y)}{\partial x} , \frac{\partial f(x, y)}{\partial y} \rbrack
\end{aligned}

Thus the gradient for this function is:

\begin{aligned}
\nabla f(x,y) = \lbrack 4x + y , x \rbrack
\end{aligned}

The gradient is extremely useful. I can take the derivative of $f(x,y)$ along a line pointing 45 degrees along the $+x$ and $+y$ directions by simply writing down the <strong>unit-vector</strong> (a vector whose magnitude is 1) that points in that direction, and taking the dot product of it with the gradient. The unit vector that points in the desired direction is:

\begin{aligned}
\hat{u} = \lbrack \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \rbrack
\end{aligned}

(check that this has a magnitude of 1). And so <strong>the derivative of $f(x,y)$ along the $\hat{u}$ direction</strong>, denoted as $\nabla_{\hat{u}} f(x,y)$, is given by:

\begin{aligned}
\nabla_{\hat{u}} f(x,y) = \hat{u} \cdot \nabla f(x,y)
\end{aligned}
\begin{aligned}
\nabla_{\hat{u}} f(x,y) = \lbrack \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \rbrack \cdot \lbrack 4x + y , x \rbrack
\end{aligned}
\begin{aligned}
\nabla_{\hat{u}} f(x,y) = \frac{4x + y}{\sqrt{2}} +  \frac{x}{\sqrt{2}} =  \frac{5x + y}{\sqrt{2}}
\end{aligned}

Thus the function $\frac{5x + y}{\sqrt{2}}$ tells us the slope of $f(x,y)$ when we are standing at the point $(x, y)$, and are facing along a line pointing 45 degrees between the $+x$ and $+y$ directions.
